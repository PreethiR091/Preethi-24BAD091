```python
# Scenario 1: E-commerce Sales Data
print("Preethi R 24BAD091")
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("data.csv", encoding='ISO-8859-1')

# Inspect dataset
print(data.head())
print(data.tail())
print(data.info())
print(data.describe())

# Check missing values
print(data.isnull().sum())

# Sales per product
sales_per_product = data.groupby("Description")["Quantity"].sum().head(10)

# Bar chart
plt.figure()
sales_per_product.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()

# Line chart
plt.figure()
sales_per_product.plot(kind='line')
plt.title("Sales Trend of Top Products")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```
