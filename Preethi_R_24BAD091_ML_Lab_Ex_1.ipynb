```python
# SCENARIO 1: E-commerce Sales Data
print("Preethi R 24BAD091")
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("data[1].csv", encoding='ISO-8859-1')

# Inspect dataset
print(data.head())
print(data.tail())
print(data.info())
print(data.describe())

# Check missing values
print(data.isnull().sum())

# Sales per product
sales_per_product = data.groupby("Description")["Quantity"].sum().head(10)

# Bar chart
plt.figure()
sales_per_product.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()

# Line chart
plt.figure()
sales_per_product.plot(kind='line')
plt.title("Sales Trend of Top Products")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
#SCENARIO 2: Hospital Patient Records
print("Preethi R 24BAD091")
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("diabetes.csv")

print(df.info())
print(df.isnull().sum())

plt.figure(figsize=(8,5))
plt.hist(df["Glucose"], bins=20, color="navy", edgecolor="black")
plt.title("Glucose Level Distribution (Histogram)")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,5))
plt.boxplot(df["Glucose"], patch_artist=True,
            boxprops=dict(facecolor="navy"),
            medianprops=dict(color="black"))
plt.title("Glucose Level Distribution (Boxplot)")
plt.ylabel("Glucose Level")
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20, color="navy", edgecolor="black")
plt.title("Age Distribution (Histogram)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,5))
plt.boxplot(df["Age"], patch_artist=True,
            boxprops=dict(facecolor="navy"),
            medianprops=dict(color="black"))
plt.title("Age Distribution (Boxplot)")
plt.ylabel("Age")
plt.show()

corr = df.corr()
plt.figure(figsize=(10,8))
plt.imshow(corr, cmap="Blues")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i, f"{corr.iloc[i, j]:.2f}",
                 ha="center", va="center", color="black", fontsize=9)

plt.title("Correlation Heatmap of Health Metrics")
plt.tight_layout()
plt.show()

```

```python
#SCENARIO 3: Housing Dataset
print("Preethi R 24BAD091")

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

housing_df = pd.read_csv("Housing.csv")

print(housing_df.head())
print(housing_df.info())
print(housing_df.describe())

print(housing_df.isnull().sum())

plt.figure(figsize=(8,6))
plt.scatter(housing_df['area'], housing_df['price'], alpha=0.6)
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Price vs Area')
plt.show()

plt.figure(figsize=(8,6))
plt.hist(housing_df['price'], bins=30, edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Number of Houses')
plt.title('Distribution of House Prices')
plt.show()

plt.figure(figsize=(8,6))
sns.boxplot(y=housing_df['price'])
plt.title('Boxplot of House Prices')
plt.show()

plt.figure(figsize=(10,8))
numeric_df = housing_df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#SCENARIO 4: Banking Customer Data
print("Preethi R 24BAD091")

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

bank_df = pd.read_csv("marketing_campaign.csv", sep="\t")

bank_df['Age'] = 2025 - bank_df['Year_Birth']

bank_df['TotalSpending'] = (
    bank_df['MntWines'] +
    bank_df['MntFruits'] +
    bank_df['MntMeatProducts'] +
    bank_df['MntFishProducts'] +
    bank_df['MntSweetProducts'] +
    bank_df['MntGoldProds']
)

print(bank_df.head())
print(bank_df.info())
print(bank_df.describe())
print(bank_df.isnull().sum())

plt.figure(figsize=(10,5))
bank_df['Age'].value_counts().sort_index().plot(kind='bar')
plt.title('Age Distribution (Bar Plot)')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(y=bank_df['Age'])
plt.title('Age Distribution (Box Plot)')
plt.show()

plt.figure(figsize=(10,5))
bank_df['Income'].dropna().astype(int).value_counts(bins=20).plot(kind='bar')
plt.title('Income Distribution (Bar Plot)')
plt.xlabel('Income Range')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(y=bank_df['Income'])
plt.title('Income Distribution (Box Plot)')
plt.show()

plt.figure(figsize=(10,5))
bank_df['TotalSpending'].value_counts(bins=20).plot(kind='bar')
plt.title('Spending Pattern (Bar Plot)')
plt.xlabel('Total Spending Range')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(y=bank_df['TotalSpending'])
plt.title('Spending Pattern (Box Plot)')
plt.show()
```
